```python
# Import the necessary libraries

import pandas as pd # to convert and read the dataframe
import mysql.connector # connect to MySQL database
from mysql.connector import Error # for error reporting
```

```python
# Create MySQL server connection

def create_server_connection(user_name, host_name, password):
    connection = None

    try:
        connection = mysql.connector.connect(
            user = user_name,
            host = host_name,
            password = password
        )
        print('MySQL server connected successfully')

    except Error as err:
        print(f'Error: {err}')

    return connection

# password
pw = '!!!!'

# Database name
db_name = 'sql_python'

conn = create_server_connection('root', 'localhost', pw)
```

```python
# Create database

def create_database(connection, db_name):
    cursor = connection.cursor()
    try:
        cursor.execute("CREATE DATABASE IF NOT EXISTS {} DEFAULT CHARACTER SET 'utf8'".format(db_name))
        print('Database created successfully')

    except Error as err:
        print(f'Error: {err}')

create_database(conn, db_name)
```

```python
# Connect to MySQL database

def create_db_connection(user, password, host, database_name):
    connection = None
    try:
        connection = mysql.connector.connect(
            user=user,
            password=password,
            host=host,
            database=database_name
        )
        print('Database successfully connected')
    except Error as err:
        print(f'Error: {err}')

    return connection
```

```python
# Execute sql queries

def execute_sql_query(connection, query):
    cursor = connection.cursor()

    try:
        cursor.execute(query)
        connection.commit()
        print('Query executed successfully')

    except Error as err:
        print(f'Error: {err}')
```

```python
# Create emloyees table

create_employees_table = """
CREATE TABLE IF NOT EXISTS employees(
emp_no int NOT NULL AUTO_INCREMENT,
DOB date NOT NULL,
first_name varchar(50) NOT NULL,
last_name varchar(50) NOT NULL,
gender enum('M', 'F') NOT NULL,
hire_date date NOT NULL,
PRIMARY KEY (emp_no)
);
"""

connection = create_db_connection('root', pw, 'localhost', db_name)
execute_sql_query(connection, create_employees_table)
```

```python
# Create salaries table

create_salary_table = """
CREATE TABLE IF NOT EXISTS salaries(
emp_no int NOT NULL AUTO_INCREMENT,
salary int(11) NOT NULL,
from_date date NOT NULL,
to_date date NOT NULL,
PRIMARY KEY (emp_no)
);
"""

connection = create_db_connection('root', pw, 'localhost', db_name)
execute_sql_query(connection, create_salary_table)
```

```python
# Load data query

def load_data_query(connection, query):
    cursor = connection.cursor()
    result = None

    try:
        cursor.execute(query)
        result = cursor.fetchall()
        print('Query executed successfully')

    except Error as err:
        print(f'Error: {err}')

    return result

sql_query = 'SELECT * FROM salaries'
connection = create_db_connection('root', pw, 'localhost', db_name)
load_data_query(connection, sql_query)
```

```python
# Run sql queries

query1 = """
SELECT * FROM salaries WHERE salary > 6000
"""

connection = create_db_connection('root', pw, 'localhost', db_name)
load_data_query(connection, query1)

# Use the guide in 'query1' to run different sql queries
```

```python
# Save data into pandas dataframe

query2 = """
SELECT * FROM salaries
"""

connection = create_db_connection('root', pw, 'localhost', db_name)
load_data_query(connection, query2)

columns = ['emp_no', 'salary', 'from_date', 'to_date']
df = pd.DataFrame(load_data_query(connection, query2), columns=columns)
df.to_csv('salaries.csv', index=False)
```
